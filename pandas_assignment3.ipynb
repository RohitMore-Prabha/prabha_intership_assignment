```python
import pandas as pd
import numpy as np
```

```python
# Create Student DataFrame with missing values and duplicate record

data = {
    "Name": ["Rohit", "Amit", "Priya", "Sneha", "Rahul", "Amit"],
    "Age": [20, np.nan, 19, 21, np.nan, np.nan],
    "Marks": [85, 78, np.nan, 90, 88, 78],
    "City": ["Pune", "Mumbai", "Nashik", "Nagpur", "Pune", "Mumbai"]
}

students = pd.DataFrame(data)

print("Original DataFrame")
students
```

```python
print("Missing Values")

students.isnull()
```

```python
print("Count of Missing Values")

students.isnull().sum()
```

```python
print("Replace Missing Values with 0")

students.fillna(0)
```

```python
print("Replace Missing Age with Average Age")

average_age = students["Age"].mean()

students["Age"] = students["Age"].fillna(average_age)

students
```

```python
print("Duplicate Records")

students.duplicated()
```

```python
print("Duplicate Rows")

students[students.duplicated()]
```

```python
print("Remove Duplicate Records")

students = students.drop_duplicates()

students
```

```python
print("DataFrame Information")

students.info()
```

```python
print("Statistical Summary")

students.describe()
```
