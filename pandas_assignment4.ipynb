```python
import pandas as pd
```

```python
# Employee Data

data = {
    "Name": ["Rohit", "Amit", "Priya", "Sneha", "Rahul", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Marketing", "HR"],
    "Salary": [45000, 38000, 55000, 42000, 60000, 40000]
}

employee = pd.DataFrame(data)

print("Employee DataFrame")
employee
```

```python
print("Salary in Ascending Order")

employee.sort_values(by="Salary", ascending=True)
```

```python
print("Salary in Descending Order")

employee.sort_values(by="Salary", ascending=False)
```

```python
print("Unique Departments")

employee["Department"].unique()
```

```python
print("Employee Count in Each Department")

employee["Department"].value_counts()
```

```python
print("Average Salary of Each Department")

employee.groupby("Department")["Salary"].mean()
```

```python
print("Maximum Salary of Each Department")

employee.groupby("Department")["Salary"].max()
```

```python
print("DataFrame Information")

employee.info()
```

```python
print("Statistical Summary")

employee.describe()
```

```python
# Pandas Assignment 4

## Theory Answers

### Q1. What is the purpose of sort_values()?

**Answer:**

The **sort_values()** method is used to sort the rows of a DataFrame based on one or more columns in ascending or descending order.

---

### Q2. What is the difference between ascending and descending sorting?

**Answer:**

- **Ascending Sorting:** Arranges data from smallest to largest (A to Z or 0 to 9).
- **Descending Sorting:** Arranges data from largest to smallest (Z to A or 9 to 0).

---

### Q3. What is the purpose of unique()?

**Answer:**

The **unique()** method returns all unique values from a column by removing duplicate values.

---

### Q4. What is the purpose of value_counts()?

**Answer:**

The **value_counts()** method counts the frequency of each unique value in a column.

---

### Q5. What is groupby()?

**Answer:**

The **groupby()** method groups rows that have the same values in a column and allows operations like **sum, mean, max, min, and count** on each group.

---

### Q6. Write the aggregation functions available in Pandas.

**Answer:**

Some commonly used aggregation functions are:

- sum()
- mean()
- max()
- min()
- count()
- median()
- std()
- var()
```

```python
print("Aggregation Functions")

print("Sum of Salary:")
print(employee["Salary"].sum())

print("\nAverage Salary:")
print(employee["Salary"].mean())

print("\nMaximum Salary:")
print(employee["Salary"].max())

print("\nMinimum Salary:")
print(employee["Salary"].min())

print("\nCount of Employees:")
print(employee["Salary"].count())
```
