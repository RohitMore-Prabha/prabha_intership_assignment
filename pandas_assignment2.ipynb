```python
import pandas as pd
```

```python
# Employee Data

data = {
    "Employee ID": [101, 102, 103, 104, 105],
    "Name": ["Rohit", "Amit", "Priya", "Sneha", "Rahul"],
    "Department": ["IT", "HR", "Finance", "IT", "Marketing"],
    "Salary": [45000, 38000, 55000, 42000, 60000],
    "City": ["Pune", "Mumbai", "Pune", "Nagpur", "Pune"]
}

employee = pd.DataFrame(data)

print("Employee DataFrame")
employee
```

```python
print("First 3 Rows")
employee.head(3)
```

```python
print("Last 2 Rows")
employee.tail(2)
```

```python
print("Dataset Shape")
print(employee.shape)
```

```python
print("Column Names")
print(employee.columns)
```

```python
print("Employees with Salary Greater Than 40000")

employee[employee["Salary"] > 40000]
```

```python
print("Employees from Pune")

employee[employee["City"] == "Pune"]
```

```python
# Pandas Assignment 2

## Theory Answers

### Q1. What is the purpose of head()?

**Answer:**

The **head()** function displays the first 5 rows of a DataFrame by default. We can specify the number of rows using `head(n)`.

---

### Q2. What is the purpose of tail()?

**Answer:**

The **tail()** function displays the last 5 rows of a DataFrame by default. We can specify the number of rows using `tail(n)`.

---

### Q3. What does shape return?

**Answer:**

The **shape** attribute returns the number of rows and columns in a DataFrame as a tuple `(rows, columns)`.

---

### Q4. What does info() display?

**Answer:**

The **info()** method displays information about the DataFrame such as:
- Number of rows and columns
- Column names
- Data types
- Non-null values
- Memory usage

---

### Q5. What does describe() display?

**Answer:**

The **describe()** method displays statistical information such as:
- Count
- Mean
- Standard Deviation
- Minimum Value
- Maximum Value
- 25%, 50%, and 75% percentiles
```
